This notebook demonstrates how to analyze and visualize queue depth telemetry data collected from a P4-based HULA load balancing system.

**Contents:**

## Section 1: P4 Telemetry Header Definition

### Telemetry Header Structure

The HULA telemetry header includes multiple path entries and a new queue_depth field for monitoring egress queue occupancy.

**Key Design Decisions:**
- **queue_depth**: 8-bit unsigned integer (0-255 bytes)
- **Placement**: After path_util3 for alignment
- **Semantics**: Egress queue occupancy at packet departure (deq_qdepth)

### P4 Header Definition Code

```p4
/* -*- P4_16 -*- */
#include <core.p4>
#include <v1model.p4>

// Type definitions
typedef bit<8>  util_t;      // Utilization: 0-255
typedef bit<24> tor_id_t;    // ToR identifier
typedef bit<48> time_t;      // Timestamp in nanoseconds

// HULA Telemetry Header with queue_depth field
header hula_t {
    // Path entry 0
    bit<24> dst_tor0;         // Destination ToR 0 identifier
    bit<8>  path_util0;       // Path utilization 0 (0-255)

    // Path entry 1
    bit<24> dst_tor1;         // Destination ToR 1 identifier
    bit<8>  path_util1;       // Path utilization 1 (0-255)

    // Path entry 2
    bit<24> dst_tor2;         // Destination ToR 2 identifier
    bit<8>  path_util2;       // Path utilization 2 (0-255)

    // Path entry 3
    bit<24> dst_tor3;         // Destination ToR 3 identifier
    bit<8>  path_util3;       // Path utilization 3 (0-255)

    // NEW: Queue depth telemetry
    bit<8>  queue_depth;      // Egress queue occupancy (0-255 bytes, saturated)
}
```

### Header Size Analysis
- **Path entries**: 4 × (24 + 8) bits = 128 bits
- **queue_depth**: 8 bits
- **Total**: 136 bits = 17 bytes
- **Overhead**: +1 byte per HULA packet

## Section 2: Adding queue_depth to P4 Data Plane

### Metadata Structure for Telemetry

```p4
struct metadata {
    bit<9>  nxt_hop;          // Next hop port
    bit<32> self_id;          // Switch identifier
    bit<32> dst_tor;          // Destination ToR
    // Note: Queue depth comes from standard_metadata, not custom metadata
}

struct headers {
    ethernet_t   ethernet;
    ipv4_t       ipv4;
    tcp_t        tcp;
    hula_t       hula;        // Includes queue_depth
}
```

## Section 3: Ingress Pipeline - Queue Depth Assignment

### Option A: Capture at Ingress (NOT RECOMMENDED)

```p4
// ❌ NOT RECOMMENDED: enq_qdepth shows queue depth at ENQUEUE
// This captures packets already in flight
action populate_queue_depth_ingress() {
    bit<32> queue_at_ingress = standard_metadata.enq_qdepth;
    hdr.hula.queue_depth = (bit<8>)(queue_at_ingress > 255 ? 255 : queue_at_ingress);
}
```

**Problem:** Doesn't show realtime congestion at egress

### Option B: Ingress Initialization

```p4
// ✅ Better: Initialize queue_depth in ingress
action init_telemetry() {
    // Initialize to 0; will be populated in egress
    hdr.hula.queue_depth = 0;
}

apply {
    if (hdr.hula.isValid()) {
        init_telemetry();     // Set initial value
    }
}
```

## Section 4: Egress Pipeline - Queue Depth Population

### ✅ RECOMMENDED: Egress Queue Depth Capture

Capture queue depth at egress **departure time** to show current congestion:

```p4
/*
 * Egress Processing: Populate queue_depth Telemetry
 * 
 * At egress, when the packet is about to be transmitted, we read
 * the queue depth and store it in the telemetry header.
 * 
 * Key Metadata:
 *   - deq_qdepth: Queue depth at packet dequeue (DEPARTURE TIME)
 *   - enq_qdepth: Queue depth at packet enqueue (not used here)
 */

control MyEgress(inout headers hdr,
                 inout metadata meta,
                 inout standard_metadata_t standard_metadata) {
    apply {
        // Only populate telemetry for HULA packets
        if (hdr.hula.isValid()) {
            // Read egress queue depth (bit<32>)
            bit<32> queue_occ = standard_metadata.deq_qdepth;
            
            // Saturating cast to bit<8>
            // If queue > 255, cap at 255 (indicates heavily congested)
            hdr.hula.queue_depth = (bit<8>)(
                queue_occ > 255 ? 255 : queue_occ
            );
        }
    }
}
```

### Type Casting Explanation

**Problem:** Direct cast causes truncation (overflow)
```p4
// ❌ WRONG: Truncates to lowest 8 bits
hdr.hula.queue_depth = (bit<8>)standard_metadata.deq_qdepth;
// If deq_qdepth = 300 (0x12C), result = 44 (0x2C) - INCORRECT!
```

**Solution:** Saturating cast with ternary operator
```p4
// ✅ CORRECT: Saturating cast
hdr.hula.queue_depth = (bit<8>)(deq_qdepth > 255 ? 255 : deq_qdepth);
// If deq_qdepth = 300, result = 255 - CORRECT!
```

## Section 5: Deparser - Telemetry Header Emission

### Deparser Configuration

```p4
/*
 * Deparser: Emit Headers
 * 
 * The deparser serializes parsed headers back into the packet.
 * Order matters for correct packet formation.
 */

control MyDeparser(packet_out packet, in headers hdr) {
    apply {
        packet.emit(hdr.ethernet);    // Layer 2
        packet.emit(hdr.ipv4);        // Layer 3
        packet.emit(hdr.tcp);         // Layer 4
        packet.emit(hdr.hula);        // ✅ Telemetry (includes queue_depth)
    }
}
```

### Header Emission Order

1. **Ethernet** (14 bytes): L2 header
2. **IPv4** (20 bytes): L3 header with protocol=0x42 (HULA)
3. **TCP** (optional, 20 bytes): L4 header
4. **HULA** (17 bytes): Telemetry header
   - Contains all path entries + **queue_depth**

**Final packet structure:**
```
Ethernet | IPv4 (proto=0x42) | HULA (incl. queue_depth) | payload
 14 B    |    20 B           |     17 B                 | N B
```

## Section 6: Receiver-Side Telemetry Parsing

Parse incoming telemetry packets using Python/Scapy.

### Key Points:
- **Protocol**: IPv4 protocol number 0x42 (66 decimal)
- **Binding**: Bind HULA header to IP layer
- **Fields**: Extract all fields including queue_depth

````python
# Receiver-side packet parsing
from scapy.all import sniff, IP, Packet, BitField, bind_layers

# Define HULA header structure for Scapy
class Hula(Packet):
    """HULA Telemetry Header with queue_depth field."""
    fields_desc = [
        # Path entry 0
        BitField("dst_tor0", 0, 24),
        BitField("path_util0", 0, 8),
        # Path entry 1
        BitField("dst_tor1", 0, 24),
        BitField("path_util1", 0, 8),
        # Path entry 2
        BitField("dst_tor2", 0, 24),
        BitField("path_util2", 0, 8),
        # Path entry 3
        BitField("dst_tor3", 0, 24),
        BitField("path_util3", 0, 8),
        # Telemetry field
        BitField("queue_depth", 0, 8),  # ← NEW: Queue depth in bytes
    ]

# Bind HULA header to IP protocol 0x42
bind_layers(IP, Hula, proto=0x42)

print("✅ HULA packet structure defined and bound to IPv4 protocol 0x42")
print(f"   Protocol number: 0x42 (66 decimal)")
print(f"   HULA header size: 17 bytes")
print(f"   Including fields: dst_tor[0-3], path_util[0-3], queue_depth")
````

### Packet Reception Handler

```python
def handle_packet(pkt, csv_writer, csv_file):
    """Handle received HULA telemetry packet."""
    import time
    
    try:
        if pkt.haslayer(Hula):
            hula = pkt[Hula]
            
            # Extract flow information
            src_ip = pkt[IP].src if pkt.haslayer(IP) else "0.0.0.0"
            dst_ip = pkt[IP].dst if pkt.haslayer(IP) else "0.0.0.0"
            src_port = pkt[TCP].sport if pkt.haslayer(TCP) else 0
            dst_port = pkt[TCP].dport if pkt.haslayer(TCP) else 0
            
            # Extract telemetry
            queue_depth = hula.queue_depth
            timestamp = time.time()
            
            # Write to CSV
            csv_writer.writerow([
                timestamp,
                src_ip,
                dst_ip,
                src_port,
                dst_port,
                queue_depth,
            ])
            csv_file.flush()
            
    except Exception as e:
        print(f"Error processing packet: {e}")
```

## Section 8: Telemetry Data Analysis and Visualization

### Load and Analyze Telemetry Data

````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Create sample telemetry data (in real scenario, load from CSV)
np.random.seed(42)
n_samples = 500

# Generate synthetic queue depth data with congestion patterns
time_s = np.linspace(0, 30, n_samples)
# Simulate congestion spikes
queue_depth = 30 + 40 * np.sin(time_s/10) + np.random.normal(0, 15, n_samples)
queue_depth = np.clip(queue_depth, 0, 255)  # Saturate at 255

# Create flow patterns
flow_ids = np.random.choice(["10.0.1.1:5000->10.0.9.1:5001",
                              "10.0.1.2:5000->10.0.9.2:5001"],
                             size=n_samples, p=[0.6, 0.4])

# Create DataFrame
df = pd.DataFrame({
    'timestamp_s': time_s,
    'src_ip': [f.split(':')[0] for f in flow_ids],
    'dst_ip': [f.split('->')[1].split(':')[0] for f in flow_ids],
    'src_port': 5000,
    'dst_port': 5001,
    'queue_depth_bytes': queue_depth.astype(int),
    'flow_id': flow_ids,
})

print("Sample Telemetry Data (first 10 rows):")
print("="*100)
print(df.head(10))
print(f"\nTotal records: {len(df)}")
print(f"Time span: {df['timestamp_s'].min():.1f}s to {df['timestamp_s'].max():.1f}s")
````

### Global Statistics

````python
# Compute global statistics
queue_depth = df['queue_depth_bytes']

stats = {
    'Count': len(queue_depth),
    'Min': queue_depth.min(),
    'Max': queue_depth.max(),
    'Mean': queue_depth.mean(),
    'Median': queue_depth.median(),
    'Std Dev': queue_depth.std(),
    'P25': queue_depth.quantile(0.25),
    'P50': queue_depth.quantile(0.50),
    'P75': queue_depth.quantile(0.75),
    'P95': queue_depth.quantile(0.95),
    'P99': queue_depth.quantile(0.99),
}

print("Global Queue Depth Statistics")
print("="*60)
for key, value in stats.items():
    if isinstance(value, float):
        print(f"{key:15s}: {value:8.2f}")
    else:
        print(f"{key:15s}: {value:8d}")
````

### Per-Flow Analysis

````python
# Per-flow statistics
flow_stats = df.groupby('flow_id')['queue_depth_bytes'].agg([
    ('Count', 'count'),
    ('Min', 'min'),
    ('Max', 'max'),
    ('Mean', 'mean'),
    ('Median', 'median'),
    ('Std Dev', 'std'),
])

print("Per-Flow Queue Depth Statistics")
print("="*100)
print(flow_stats.to_string())
````

### Visualization: Queue Depth Timeline

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)

# Create figure
fig, ax = plt.subplots(figsize=(14, 6))

# Plot queue depth over time for each flow
for flow_id in df['flow_id'].unique():
    flow_data = df[df['flow_id'] == flow_id]
    ax.plot(flow_data['timestamp_s'], flow_data['queue_depth_bytes'],
            label=flow_id, linewidth=2, alpha=0.8)

# Add rolling average
rolling_avg = df.groupby('timestamp_s')['queue_depth_bytes'].mean().rolling(window=20).mean()
ax.plot(rolling_avg.index, rolling_avg.values,
        label='Rolling Avg (window=20)', linestyle='--', linewidth=2.5, color='red')

# Formatting
ax.set_xlabel('Time (seconds)', fontsize=12)
ax.set_ylabel('Queue Depth (bytes)', fontsize=12)
ax.set_title('HULA Queue Depth Over Time', fontsize=14, fontweight='bold')
ax.legend(loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)
ax.set_ylim(0, 260)

# Add congestion threshold
ax.axhline(y=150, color='orange', linestyle=':', linewidth=2, label='Moderate Congestion (150 B)', alpha=0.7)
ax.axhline(y=200, color='red', linestyle=':', linewidth=2, label='Heavy Congestion (200 B)', alpha=0.7)

plt.tight_layout()
plt.show()

print("✅ Queue Depth Timeline Visualization Complete")
````

### Visualization: Queue Depth Distribution

````python
# Queue depth distribution
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Histogram of queue depth
ax = axes[0, 0]
ax.hist(df['queue_depth_bytes'], bins=30, edgecolor='black', alpha=0.7, color='skyblue')
ax.axvline(df['queue_depth_bytes'].mean(), color='red', linestyle='--',
           linewidth=2, label=f'Mean: {df["queue_depth_bytes"].mean():.1f}')
ax.set_xlabel('Queue Depth (bytes)')
ax.set_ylabel('Frequency')
ax.set_title('Queue Depth Distribution')
ax.legend()
ax.grid(True, alpha=0.3)

# 2. Box plot per flow
ax = axes[0, 1]
df.boxplot(column='queue_depth_bytes', by='flow_id', ax=ax)
ax.set_xlabel('Flow')
ax.set_ylabel('Queue Depth (bytes)')
ax.set_title('Queue Depth Distribution per Flow')
ax.get_figure().suptitle('')  # Remove automatic title
plt.sca(ax)
plt.xticks(rotation=45, ha='right')

# 3. CDF (Cumulative Distribution Function)
ax = axes[1, 0]
sorted_qd = np.sort(df['queue_depth_bytes'])
y = np.arange(1, len(sorted_qd) + 1) / len(sorted_qd)
ax.plot(sorted_qd, y, linewidth=2, color='darkblue')
ax.axvline(df['queue_depth_bytes'].quantile(0.95), color='orange', linestyle='--',
           linewidth=2, label='P95')
ax.axvline(df['queue_depth_bytes'].quantile(0.99), color='red', linestyle='--',
           linewidth=2, label='P99')
ax.set_xlabel('Queue Depth (bytes)')
ax.set_ylabel('Cumulative Probability')
ax.set_title('Cumulative Distribution Function (CDF)')
ax.legend()
ax.grid(True, alpha=0.3)

# 4. Congestion level breakdown
ax = axes[1, 1]
congestion_levels = pd.cut(df['queue_depth_bytes'],
                            bins=[0, 50, 100, 150, 200, 256],
                            labels=['Empty', 'Low', 'Medium', 'High', 'Critical'])
congestion_counts = congestion_levels.value_counts().sort_index()
colors = ['green', 'yellow', 'orange', 'red', 'darkred']
ax.bar(range(len(congestion_counts)), congestion_counts.values, color=colors, edgecolor='black')
ax.set_xticks(range(len(congestion_counts)))
ax.set_xticklabels(congestion_counts.index)
ax.set_ylabel('Count')
ax.set_title('Queue Depth Congestion Levels')
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("✅ Queue Depth Distribution Visualization Complete")
````

### Congestion Detection

````python
def detect_congestion_periods(df, threshold=150, min_duration=0.1):
    """
    Detect periods of congestion (queue_depth > threshold).
    
    Args:
        df: DataFrame with telemetry data
        threshold: Queue depth threshold (bytes)
        min_duration: Minimum duration to count as congestion period (seconds)
    
    Returns:
        List of congestion periods with start, end, peak, duration
    """
    congestion_mask = df['queue_depth_bytes'] > threshold
    congestion_changes = congestion_mask.astype(int).diff().nonzero()[0]
    
    periods = []
    start_idx = None
    
    for idx in congestion_changes:
        if congestion_mask.iloc[idx]:
            start_idx = idx
        elif start_idx is not None:
            duration = df.iloc[idx]['timestamp_s'] - df.iloc[start_idx]['timestamp_s']
            if duration >= min_duration:
                peak = df.iloc[start_idx:idx]['queue_depth_bytes'].max()
                periods.append({
                    'start_time': df.iloc[start_idx]['timestamp_s'],
                    'end_time': df.iloc[idx]['timestamp_s'],
                    'duration': duration,
                    'peak': peak,
                })
            start_idx = None
    
    return periods

# Detect congestion periods
threshold = 150
congestion_periods = detect_congestion_periods(df, threshold=threshold)

print(f"Congestion Detection (threshold: {threshold} bytes)")
print("="*80)
if congestion_periods:
    print(f"Found {len(congestion_periods)} congestion period(s):\n")
    for i, period in enumerate(congestion_periods, 1):
        print(f"  Period {i}:")
        print(f"    Start:    {period['start_time']:.2f}s")
        print(f"    End:      {period['end_time']:.2f}s")
        print(f"    Duration: {period['duration']:.2f}s")
        print(f"    Peak:     {period['peak']} bytes")
        print()
else:
    print(f"✅ No congestion periods detected (threshold: {threshold} bytes)")
````

## Section 9: Type Casting and Debugging Solutions

### Error 1: "Cannot cast implicitly type 'bit<8>'"

#### Problem

```p4
// ❌ WRONG: P4 compiler error
hdr.hula.queue_depth = standard_metadata.deq_qdepth;  // Type mismatch!
// Error: Cannot assign bit<32> to bit<8> without explicit cast
```

#### Solution: Explicit Cast

```p4
// ✅ CORRECT: Explicit cast with type annotation
hdr.hula.queue_depth = (bit<8>)standard_metadata.deq_qdepth;
```

**Warning:** This truncates to lowest 8 bits, which causes overflow if deq_qdepth > 255

### Error 2: "Incorrect Value (Truncation/Overflow)"

````python
# Demonstrate truncation issue
print("Type Casting Issue: Truncation vs Saturation")
print("="*80)

test_values = [0, 100, 255, 256, 300, 500, 1000]

print(f"{'Value':>6} | {'Binary':>12} | {'Truncated (bit<8>)':>20} | {'Saturated (bit<8>)':>20}")
print("-" * 80)

for val in test_values:
    truncated = val & 0xFF  # Keeps only lowest 8 bits
    saturated = min(val, 255)
    
    print(f"{val:>6} | {bin(val):>12} | {truncated:>20} | {saturated:>20}")

print("\n❌ Truncation: Direct cast loses information for values > 255")
print("✅ Saturation: Cap at 255 preserves 'heavily congested' information")
````

### Error 3: "Queue Depth Always 0"

#### Possible Causes and Solutions

| Cause | Solution |
|-------|----------|
| Reading in **Ingress** instead of **Egress** | Move logic to `MyEgress` control block |
| Using `enq_qdepth` (enqueue depth) | Use `deq_qdepth` (departure time depth) |
| HULA header not valid | Check: `if (hdr.hula.isValid())` before writing |
| Deparser not emitting HULA header | Verify: `packet.emit(hdr.hula);` in deparser |
| No traffic flowing | Generate traffic: `iperf -c <dst> -b 50M` |

#### Debugging Steps

```bash
# 1. Verify HULA packets are being sent
tcpdump -i eth0 'proto 66'

# 2. Dump full packet structure
tcpdump -i eth0 -XX 'proto 66'

# 3. Parse packet with Scapy
python3 -c "
from scapy.all import IP, sniff, Packet, BitField, bind_layers

class Hula(Packet):
    fields_desc = [BitField('queue_depth', 0, 8)]

bind_layers(IP, Hula, proto=0x42)
sniff(iface='eth0', prn=lambda x: print(x.show()) if x.haslayer(Hula) else None)
"
```

````python
# Validation script
import os
import sys

def validate_setup():
    """
    Validate HULA queue_depth telemetry setup.
    """
    checks = []
    
    # Check 1: P4 files exist
    p4_file = "switch.p4"
    if os.path.exists(p4_file):
        with open(p4_file, 'r') as f:
            content = f.read()
            has_queue_depth = 'queue_depth' in content and 'bit<8>' in content
            has_egress_logic = 'MyEgress' in content and 'deq_qdepth' in content
            has_deparser = 'MyDeparser' in content and 'packet.emit(hdr.hula)' in content
        
        checks.append(("P4 has queue_depth field", has_queue_depth))
        checks.append(("Egress has deq_qdepth logic", has_egress_logic))
        checks.append(("Deparser emits HULA", has_deparser))
    
    # Check 2: Receiver script
    receiver = "test-scripts/telemetry_receiver.py"
    if os.path.exists(receiver):
        with open(receiver, 'r') as f:
            content = f.read()
            has_hula_class = 'class Hula' in content
            has_queue_depth_field = 'BitField("queue_depth"' in content
            has_csv_write = 'csv' in content and 'writerow' in content
        
        checks.append(("Receiver has Hula class", has_hula_class))
        checks.append(("Receiver includes queue_depth", has_queue_depth_field))
        checks.append(("Receiver writes to CSV", has_csv_write))
    
    # Print results
    print("HULA Queue Depth Setup Validation")
    print("="*60)
    for check_name, result in checks:
        status = "✅" if result else "❌"
        print(f"{status} {check_name}")
    
    all_passed = all(result for _, result in checks)
    if all_passed:
        print("\n✅ All checks passed! Setup is ready.")
    else:
        print("\n❌ Some checks failed. Review the P4 code and receiver script.")
    
    return all_passed

# Note: This would only work if run in the actual project directory
print("Validation function defined. Run this when in the project directory.")
````

### Type Casting Key Points

```p4
// ❌ Wrong: Truncation
hdr.hula.queue_depth = (bit<8>)standard_metadata.deq_qdepth;

// ✅ Correct: Saturating cast
hdr.hula.queue_depth = (bit<8>)(
    deq_qdepth > 255 ? 255 : deq_qdepth
);
```
